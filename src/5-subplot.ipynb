### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt

# Data for the plots
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
categories = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create 2x2 subplot layout
fig, axs = plt.subplots(2, 2)

# Plot a line chart in the first subplot
axs[0, 0].plot(x, y, marker='o')
axs[0, 0].set_title('Line Chart')

# Plot a bar chart in the second subplot
axs[0, 1].bar(categories, sizes)
axs[0, 1].set_title('Bar Chart')

# Plot a scatter plot in the third subplot
axs[1, 0].scatter(x, y, c='red')
for xi, yi in zip(x, y):
    axs[1, 0].annotate(f'({xi}, {yi})', (xi, yi), textcoords="offset points", xytext=(0, 10), ha='center')
axs[1, 0].set_title('Scatter Plot')

# Plot a pie chart in the fourth subplot
axs[1, 1].pie(sizes, labels=categories, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Display the subplot layout
plt.show()
```
